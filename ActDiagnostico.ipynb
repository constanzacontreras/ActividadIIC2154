 Pregunta 1)

```python
def top_retweet(cantidad, df):
    df_orden = df.sort_values(by = 'retweetCount', ascending=False)
    return df_orden.head(cantidad)    
```

Pregunta 2)

```python
def top_users(cantidad, df):
    user = df['user']
    usernames = []
    for key in df['user']:
        usernames.append(key['username'])
    df['username'] = usernames
    
    freq = df.groupby(['username']).count() 
    top_2 = freq.sort_values(by = 'url', ascending=False)
    return top_2.head(cantidad)
```

Pregunta 3)

```python
def top_days(cantidad,df):
    df['date'] = pd.to_datetime(df['date']).dt.date
    freq_dates = df.groupby(['date']).count() 
    df_orden_dates = freq_dates.sort_values(by = 'url', ascending=False)
    return df_orden_dates.head(cantidad) 
```

Pregunta 4

```python
def top_hash(cantidad, df):
    hashtags=df['renderedContent'].str.extractall(r"(#\S+)")
    freqs_hash = hashtags[0].value_counts()
    return freqs_hash[:cantidad]
```

```python
def main(request):
    #pregunta 1)
    if (request == 1):
        return top_retweet(10,data_json)
    
    
    #pregunta 2)
    if (request == 2):
        return top_users(10,data_json)
    

    
    #pregunta 3)
    if (request == 3):
        return top_days(10,data_json )
    

    
    #pregunta 4)
    if (request == 4):
        return top_hash(10, data_json)
    
    else:
        return 0
```
